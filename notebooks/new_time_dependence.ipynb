```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
time_dependence_file = "/Users/alexanderescamilla/Desktop/time_dependence_data.csv"
df = pd.read_csv(time_dependence_file)
```

```python
df
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(df['Simulation Time (Myr)'], df["$\Sigma_{gas}$ = 100"], label = r" $\Sigma = 100 $", c='blue', alpha=0.7)
ax.plot(df['Simulation Time (Myr)'], df["$\Sigma_{gas}$ = 500"], label = r" $\Sigma = 500 $", c='green', alpha=0.7)
ax.plot(df['Simulation Time (Myr)'], df["$\Sigma_{gas}$ = 1000"], label =  r" $\Sigma = 1000 $", c='black', alpha=0.6)
ax.legend(fontsize=15)
ax.set_xlabel(r" Simulation Time (Myr)", fontsize= 25);
ax.set_ylabel(r"Log $\epsilon_{\rm ff}$", fontsize=25);
ax.tick_params(axis="x", labelsize=16)
ax.tick_params(axis="y", labelsize=16)
#ax.set_title(r"Log $\epsilon_{ff}$ vs. t (Myr) at Different Cutoffs", fontsize=20);
pos = ax.get_position()
ax.set_position([pos.x0, pos.y0, pos.width * 0.9, pos.height])


#plt.savefig('test', facecolor='white', transparent=False)
```
