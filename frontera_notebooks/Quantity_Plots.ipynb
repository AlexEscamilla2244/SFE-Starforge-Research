```python

def quantity_plots(files):
    fig, axs = plt.subplots(5, figsize=(12,20))
    for file in files:
        with h5py.File(str(file), 'r') as F:
            Area = (F["Area"][:])
            sigma_gas = np.linspace(65,2500,2435)
            new_name = file.split("/")[-1].replace(".quantities.hdf5", "")
       
            axs[0].plot(np.log10(sigma_gas), np.log10(Area), label= new_name)
            axs[0].set_ylabel(r"Area", fontsize = 15)
            axs[0].legend(loc='center left', bbox_to_anchor=(1, 0.5))
    for file in files:
        with h5py.File(str(file), 'r') as F:
            M_Gas = (F["M_Gas"][:])
            sigma_gas = np.linspace(65,2500,2435)
            new_name = file.split("/")[-1].replace(".quantities.hdf5", "")
            axs[1].plot(np.log10(sigma_gas), np.log10(M_Gas), label= new_name)
        
            axs[1].set_ylabel(r"Mass of Gas $(\rm M_\odot\,pc^{-2})$", fontsize = 15)
            axs[1].legend(loc='center left', bbox_to_anchor=(1, 0.5))
    for file in files:
        with h5py.File(str(file), 'r') as F:
            NYSOs = (F["NYSOs"][:])
            sigma_gas = np.linspace(65,2500,2435)
            new_name = file.split("/")[-1].replace(".quantities.hdf5", "")
       
            axs[2].plot(np.log10(sigma_gas), (NYSOs), label = new_name)
            #axs[2].set_xlabel(r"Log $\Sigma_{gas}$ $(\rm M_\odot\,pc^{-2})$", fontsize = 15)
            axs[2].set_ylabel(r"Number of Young Stellar Objects (YSOs)", fontsize = 15)
            axs[2].legend(loc='center left', bbox_to_anchor=(1, 0.5))
    for file in files:
        with h5py.File(str(file), 'r') as F:
            T_ff = (F["T_ff"][:])
            sigma_gas = np.linspace(65,2500,2435)
            new_name = file.split("/")[-1].replace(".quantities.hdf5", "")
            #axs[3].set_xlabel(r"Log $\Sigma_{gas}$ $(\rm M_\odot\,pc^{-2})$", fontsize = 15)
            axs[3].set_ylabel(r"Free Fall Time $(T_{ff})$", fontsize = 15)
            axs[3].plot(np.log10(sigma_gas), (T_ff), label=new_name)
            axs[3].legend(loc='center left', bbox_to_anchor=(1, 0.5))

    for file in files:
        with h5py.File(str(file), 'r') as F:
            T_ff = (F["T_ff"][:])
            M_Gas = (F["M_Gas"][:])
            sigma_gas = np.linspace(65,2500,2435)
            new_name = file.split("/")[-1].replace(".quantities.hdf5", "")
           
            axs[4].plot(np.log10(sigma_gas), np.log10(M_Gas/T_ff), label=new_name)
        
        
            axs[4].set_xlabel(r"Log $\Sigma_{gas}$ $(\rm M_\odot\,pc^{-2})$", fontsize = 15)
            axs[4].set_ylabel(r"log $M_{gas} / T_{ff}$", fontsize = 15)
            axs[4].legend(loc='center left', bbox_to_anchor=(1, 0.5))

fig.suptitle(r"Log $\Sigma_{gas}$ $(\rm M_\odot\,pc^{-2})$ vs. Distinct Quantities", fontsize=20 );
#fig.savefig('quantities', facecolor='white', transparent=False, bbox_inches="tight")
```
